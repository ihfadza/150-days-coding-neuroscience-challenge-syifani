# Day 4: PecahKata

```python
import pygame
import pyttsx3
import random
import sys

# Init pygame & TTS
pygame.init()
engine = pyttsx3.init()

# Screen setup
WIDTH, HEIGHT = 800, 400
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)
GREEN = (100, 200, 100)
RED = (255, 100, 100)
BLACK = (0, 0, 0)
FONT = pygame.font.SysFont('arial', 40)

screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("PecahKata – Game Fonik")

# Kata dan fonem
words = [
    {"word": "kucing", "syllables": ["ku", "cing"]},
    {"word": "rumah", "syllables": ["ru", "mah"]},
    {"word": "bola", "syllables": ["bo", "la"]},
    {"word": "meja", "syllables": ["me", "ja"]},
    {"word": "topi", "syllables": ["to", "pi"]},
]

# Game state
current_word = random.choice(words)
current_index = 0
score = 0
feedback = ""

# Render text
def draw_text(text, x, y, color=BLACK):
    img = FONT.render(text, True, color)
    screen.blit(img, (x, y))

# TTS
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Button object
class SyllableButton:
    def __init__(self, text, x, y, w=150, h=80):
        self.text = text
        self.rect = pygame.Rect(x, y, w, h)
        self.color = GRAY
        self.clicked = False
        self.correct = False

    def draw(self, surf):
        # Warna berdasarkan status klik
        if self.clicked:
            self.color = GREEN if self.correct else RED
        else:
            self.color = GRAY

        pygame.draw.rect(surf, self.color, self.rect, border_radius=10)
        draw_text(self.text, self.rect.x + 40, self.rect.y + 20)

    def is_clicked(self, pos):
        return self.rect.collidepoint(pos)

def new_round():
    global current_word, current_index, feedback
    current_word = random.choice(words)
    current_index = 0
    feedback = ""
    create_buttons()

def create_buttons():
    global buttons
    buttons = []
    sylls = current_word['syllables']
    x_start = (WIDTH - (len(sylls) * 160)) // 2
    for i, syll in enumerate(sylls):
        btn = SyllableButton(syll, x_start + i * 160, 200)
        buttons.append(btn)

create_buttons()
running = True

# Main loop
while running:
    screen.fill(WHITE)

    # Judul & skor
    draw_text(f"Tebak: {current_word['word']}", 50, 50)
    draw_text(f"Skor: {score}", WIDTH - 180, 50)

    # Feedback
    draw_text(feedback, WIDTH // 2 - 100, 320, RED if "Salah" in feedback else GREEN)

    # Draw buttons
    for btn in buttons:
        btn.draw(screen)

    pygame.display.flip()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            break
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pos = pygame.mouse.get_pos()
            for btn in buttons:
                if btn.is_clicked(pos):
                    btn.clicked = True
                    if btn.text == current_word["syllables"][current_index]:
                        btn.correct = True
                        speak(btn.text)
                        current_index += 1
                        feedback = "Benar!"
                        if current_index == len(current_word["syllables"]):
                            score += 1
                            feedback = "✅ Skor +1!"
                            pygame.time.delay(500)
                            new_round()
                    else:
                        btn.correct = False
                        feedback = "❌ Salah! Ulang ya"
                        current_index = 0
                        # Reset semua tombol
                        for b in buttons:
                            b.clicked = False
                            b.correct = False

pygame.quit()
sys.exit()
```
